```python
ax = sns.regplot(x=pop_vs_under_the_area_final_arima['April 1, 2010 - Census'], y=pop_vs_under_the_area_final_arima['AUC'])
ax.set(xlabel="Population (April 1, 2010)", ylabel="AUB")
ax.set_title("AUB Algorithm Output vs. Population")
slope, intercept, r_value, p_value, std_err = sy.linregress(pop_vs_under_the_area_final_arima['AUC'], pop_vs_under_the_area_final_arima['April 1, 2010 - Census'])
print("r = " + str(r_value))
plt.show()
```
