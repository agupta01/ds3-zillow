```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metro_data = pd.read_csv('/Users/camilledunning/Downloads/zecon/Metro_time_series.csv')
county_crosswalk = pd.read_csv('/Users/camilledunning/Downloads/zecon/CountyCrosswalk_Zillow.csv')

cc_unique = county_crosswalk[['MetroName_Zillow', 'CBSACode']].dropna().drop_duplicates()
cc_unique['CBSACode'] = cc_unique['CBSACode'].astype(int)
```

```python
metro_name = cc_unique['MetroName_Zillow'].to_numpy()
cbsa_code = cc_unique['CBSACode'].to_numpy()
kv_pairs = dict(zip(cbsa_code, metro_name))

def find_values(search_key):
    v = 0
    for k, v in kv_pairs.items():
        if k == search_key:
            v = kv_pairs[k]
            break
    return v

metro_data = metro_data.drop(metro_data[metro_data['RegionName'] == 'United_States'].index)
reg_values = []

for c in metro_data['RegionName']:
    reg_values = np.append(reg_values, find_values(int(c)))

metro_data['MetroName'] = reg_values
metro_data
```
