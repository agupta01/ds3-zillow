```python
AUB_norm_path = 'datasets/AUB_results_normalized.csv'
```

```python
AUB_norm_df = pd.read_csv(AUB_norm_path)
```

```python
AUB_norm_df = AUB_norm_df.drop('Unnamed: 0', axis= 1)
```

```python
AUB_norm_df = AUB_norm_df.rename(columns={'CBSA_Code': 'CBSACode'})
```

```python
AUB_norm_df = AUB_norm_df.rename(columns={'AUC': 'AUC Nomalized'})
```

```python
AUB_norm_df = AUB_norm_df.rename(columns={'AUC Nomalized': 'AUC Normalized'})
```

```python
AUB_vs_unemployment_final = pd.merge(AUB_vs_unemployment_final, AUB_norm_df, how='left')
```

```python
AUB_vs_unemployment_final
```

```python
# print('2008')
ax = sns.regplot(x=AUB_vs_unemployment_final['AUC Normalized'], y=AUB_vs_unemployment_final['Unemployment Rate'], fit_reg=True, ci=75)
ax.set(xlabel="AUB (Normalized)", ylabel="Unemployment Rate (2008)")
ax.set_title("Unemployment Rate vs. AUB Algorithm Output")
import scipy.stats as sy
slope, intercept, r_value, p_value, std_err = sy.linregress(AUB_vs_unemployment_final['AUC Normalized'], AUB_vs_unemployment_final['Unemployment Rate'])
print("r = " + str(r_value))
plt.show()
```
